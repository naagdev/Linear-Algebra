```
pool = img_cvt[:,:,0]
```

```
np.linalg.matrix_rank(pool)#Reduced rank matrix
```

```
tranpose = pool.T
SM = pool@tranpose #square matrix
inv_SM = np.linalg.inv(SM) # inverse of square matrix
LR = tranpose@inv_SM # Right Inverse
I = (pool@tranpose)@inv_SM #Identity matrix
```
